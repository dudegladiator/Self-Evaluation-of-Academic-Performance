Correlation between GPA and other variables

```python
import pandas as pd
```

Load the AllSlopes.csv file

```python
AllSlopes=pd.read_csv("/content/AllSlopes.csv")
AllSlopes.set_index('uids', inplace=True)
```

```python
AllSlopes.columns.tolist()
```

```python
AllSlopes
```

```python
Correlation=pd.DataFrame()
```

```python
for i,columns in enumerate(AllSlopes.columns.tolist()):
  if i>1:
    Correlation.at[0,columns]=AllSlopes['gpa_all'].corr(AllSlopes[columns])
    Correlation.at[2,columns]=AllSlopes['Spring_gpa'].corr(AllSlopes[columns])
    print("GPA ALL -  " +str(columns)+":  "+str(AllSlopes['gpa_all'].corr(AllSlopes[columns])))
    print("Spring GPA - "+str(columns)+":  "+str(AllSlopes['Spring_gpa'].corr(AllSlopes[columns])))
```

```python
Correlation['Index']=['GPA all','Spring GPA']
Correlation.set_index('Index', inplace=True)
Correlation
```

```python
Correlation.to_csv("AllCorrelation.csv",index=False)
```
