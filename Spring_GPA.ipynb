```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
import datetime
```

```python
from scipy.stats import ttest_ind
```

```python
AllSlopes=pd.read_csv("/content/AllSlopes.csv")
AllSlopes.set_index('uids', inplace=True)
```

```python
columns=AllSlopes.columns.tolist()
for i,column in enumerate(columns):
  if i>1:
    if(abs(AllSlopes['Spring_gpa'].corr(AllSlopes[column]))>0.20 and ttest_ind(AllSlopes[column],AllSlopes['Spring_gpa']).pvalue<0.05):
      print(str(column)+":  "+str(AllSlopes['Spring_gpa'].corr(AllSlopes[column])))
      print(str(column)+":  "+str(ttest_ind(AllSlopes[column],AllSlopes['Spring_gpa']).pvalue))
    else :
      AllSlopes.drop(column,axis=1,inplace=True)
```

```python
AllSlopes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
allSlopes = AllSlopes.fillna(10)
X=allSlopes.drop(['gpa_all','Spring_gpa'],axis=1)
y=allSlopes['Spring_gpa']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=75)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create a Lasso regression model
lasso_model = Lasso(alpha=0.04)  # Set the regularization strength (alpha)
0
# Train the model
lasso_model.fit(X_train_scaled, y_train)
```

```python
# Make predictions
y_pred = lasso_model.predict(X_test_scaled)

# Calculate Mean Squared Error (MSE) to evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

mae = mean_absolute_error(y_test,y_pred)
print("Mean Absolute Error:", mae)

# Print the coefficients of the features
print("Coefficients:", lasso_model.coef_)
```
